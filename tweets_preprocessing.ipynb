```python
import pandas as pd

df = pd.read_csv("./data/corpora/tweets_raw/tweets.csv", encoding = "cp1252", header=None)
df.head()
```

```python
#remove the rows containing the mentions
df_tweets = df[ df[5].str.contains( "@" ) ==False ]
df_tweets
```

```python
df_tweets = df_tweets.iloc[0:5000,5:6]
df_tweets
```

```python
#iterate over a series
for index, value in df_tweets[5].items():
    path = "./data/corpora/tweets/tweet"+str(index)
    with open(path, "x") as reviewFile:
        reviewFile.write(value)
        reviewFile.close()
```
