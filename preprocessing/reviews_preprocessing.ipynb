```python
import pandas as pd

df = pd.read_csv("./data/corpora/reviews_raw/reviews.csv")

df
```

```python
trutful_df = df.loc[df["deceptive"] == "truthful"]
trutful_df = trutful_df["text"]
#iterate over a series
for index, value in trutful_df.items():
    path = "./data/corpora/trutful_reviews/truthful"+str(index)
    with open(path, "x") as reviewFile:
        reviewFile.write(value)
        reviewFile.close()
```

```python
deceptive_df = df.loc[df["deceptive"] == "deceptive"]
deceptive_df = deceptive_df["text"]
#iterate over a series
for index, value in deceptive_df.items():
    path = "./data/corpora/deceptive_reviews/deceptive"+str(index)
    with open(path, "x") as reviewFile:
        reviewFile.write(value)
        reviewFile.close()
```
